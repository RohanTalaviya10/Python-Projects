```python
import requests
from PIL import Image
from io import BytesIO

def download_dish_image(dish_name,no_of_images):
    API_KEY = 'AIzaSyBY9FNTa3PAmai0ipY_BajuX1W4dgVUGzE'  # Replace with your API key
    CX = 'c0a8c039e9bb44fd0'  # Replace with your Search Engine ID
    SEARCH_URL = "https://www.googleapis.com/customsearch/v1"

    headers = {
        "User-Agent": (
            "MyFoodImageDownloader/1.0 (https://mywebsite.com; "
            "support@mywebsite.com)"
        )
    }

    params = {
        "q": dish_name +  " with full HD and plain plate bakground",
        "cx": CX,
        "key": API_KEY,
        "searchType": "image",
        "num": no_of_images  # Number of images to fetch
    }

    try:
        # Search for the dish
        response = requests.get(SEARCH_URL, params=params, headers=headers)
        response.raise_for_status()
        data = response.json()

        print(data)

        # Get the image links and download each
        if "items" in data:
            for number in range(no_of_images):
                img_url = data['items'][number]['link']


                # Request the image with headers
                img_response = requests.get(img_url, headers=headers)
                img_response.raise_for_status()

                # Open the image from the response
                image = Image.open(BytesIO(img_response.content))

                # Resize the image to 120 x 120 pixels
                resized_image = image.resize((128, 128))


                # Save the resized image
                filename = f"{dish_name.replace(' ', '_')}_{number}.jpg"
                resized_image.save(filename)

                print(f"Image downloaded and resized successfully: {filename}")
        else:
            print("No image found for the dish.")

    except Exception as e:
        print(f"Error occurred: {e}")
```
