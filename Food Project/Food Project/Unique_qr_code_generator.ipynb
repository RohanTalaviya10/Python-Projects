```python
import qrcode
from PIL import Image, ImageDraw, ImageFont

def generate_qr_with_text(data, text, output_path):
    # Generate QR Code
    qr = qrcode.QRCode(version=1, box_size=10, border=4)
    qr.add_data(data)
    qr.make(fit=True)
    qr_img = qr.make_image(fill_color="black", back_color="white").convert("RGB")

    # Set up font (use a TTF font for better control over size)
    try:
        font = ImageFont.truetype("/content/Arial.ttf", size=30)  # Ensure 'arial.ttf' is available
    except OSError:
        print("Font not found. Using default font.")
        font = ImageFont.load_default()

    # Calculate text position
    qr_width, qr_height = qr_img.size
    draw = ImageDraw.Draw(qr_img)

    # Use `textbbox` for accurate dimensions
    text_bbox = draw.textbbox((0, 0), text, font=font)
    text_width = text_bbox[2] - text_bbox[0]
    text_height = text_bbox[3] - text_bbox[1]

    # Create a new image with enough space for text
    padding = 30  # Additional padding to prevent text cutting off
    new_height = qr_height + text_height + padding + 50
    combined_img = Image.new("RGB", (qr_width, new_height), "white")
    combined_img.paste(qr_img, (0, 0))

    # Draw text below the QR code
    draw = ImageDraw.Draw(combined_img)
    text_position = ((qr_width - text_width) // 2, qr_height + (padding // 2))
    draw.text(text_position, text, fill="black", font=font)

    # Save the image
    combined_img.save(output_path)
    print(f"QR Code with text saved to {output_path}")

# Example usage
generate_qr_with_text("https://example.com", "Table 1", "qr_with_text_below.png")
```
