```python
def calculate_energy_distribution(data):

    # Extract nutrient values
    total_energy = next(nutrient for nutrient in data["nutrients"]["macro_nutrients"] if nutrient["name"] == "energy")["value"]  # Total energy in kcal

    unit_conversion = {
      "g": 1,        # 1 gram = 1 gram
      "kg": 1000,    # 1 kilogram = 1000 grams
      "mg": 0.001,   # 1 milligram = 0.001 grams
      "µg": 0.000001, # 1 microgram = 0.000001 grams
    }

    # Nutrient quantities in grams
    protein = next(nutrient for nutrient in data["nutrients"]["macro_nutrients"] if nutrient["name"] == "proteins")["value"]
    protein_unit = next(nutrient for nutrient in data["nutrients"]["macro_nutrients"] if nutrient["name"] == "proteins")["unit"]

    carbs = next(nutrient for nutrient in data["nutrients"]["macro_nutrients"] if nutrient["name"] == "carbs")["value"]
    carbs_unit = next(nutrient for nutrient in data["nutrients"]["macro_nutrients"] if nutrient["name"] == "carbs")["unit"]

    fats = next(nutrient for nutrient in data["nutrients"]["macro_nutrients"] if nutrient["name"] == "fats")["value"]
    fats_unit = next(nutrient for nutrient in data["nutrients"]["macro_nutrients"] if nutrient["name"] == "fats")["unit"]

    fiber = next(nutrient for nutrient in data["nutrients"]["macro_nutrients"] if nutrient["name"] == "fibers")["value"]
    fiber_unit = next(nutrient for nutrient in data["nutrients"]["macro_nutrients"] if nutrient["name"] == "fibers")["unit"]

    # Caloric values per gram for each nutrient
    caloric_values = {
        "protein": 4,  # kcal per gram
        "carbs": 4,   # kcal per gram
        "fats": 9,    # kcal per gram
        "fiber": 2    # kcal per gram
    }


    # Calculate energy contribution for each nutrient
    energy_contributions = {
        "protein": (protein * unit_conversion.get(protein_unit, 1)) * caloric_values["protein"],
        "carbs": (carbs * unit_conversion.get(carbs_unit, 1)) * caloric_values["carbs"],
        "fats": (fats * unit_conversion.get(fats_unit, 1)) * caloric_values["fats"],
        "fiber": (fiber * unit_conversion.get(fiber_unit, 1)) * caloric_values["fiber"]
    }

    # Add total energy from protein, carbs, fats, and fiber
    total_nutrient_energy = sum(energy_contributions.values())

    # Calculate the percentage difference between total_nutrient_energy and total_energy
    total_energy_max = total_energy if total_nutrient_energy <= total_energy  else total_nutrient_energy
    energy_difference_percentage = ((total_nutrient_energy - total_energy) / total_energy_max) * 100

    if energy_difference_percentage >= 5:
        total_energy = total_nutrient_energy

    # Calculate percentage contribution for each nutrient
    percentage_contributions = {}
    for nutrient, energy in energy_contributions.items():
        percentage = (energy / total_energy) * 100
        percentage_contributions[nutrient] = percentage


    return percentage_contributions



percentages = calculate_energy_distribution(data)
print("Energy Contribution (%):", percentages)
```
