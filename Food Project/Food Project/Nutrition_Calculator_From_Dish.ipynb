```python
# Nutrient requirement function


def calculate_nutrient_requirements(daily_requirement, nutrient_name, max_limit=None):
    # Percentage ranges for different meals
    allocation_percentage = {
        "Breakfast": (0.20, 0.25),
        "Lunch": (0.30, 0.35),
        "Snack": (0.10, 0.15),
        "Dinner": (0.30, 0.35),
    }

    # Calculate the per-meal requirement
    per_meal_requirements = {}
    for meal, (min_percentage, max_percentage) in allocation_percentage.items():
        min_requirement = daily_requirement * min_percentage
        max_requirement = daily_requirement * max_percentage

        per_meal_requirements["Nutrient Name"] = nutrient_name

        # Apply upper limit if provided
        if max_limit:
            per_meal_requirements[meal] = {
                "Min Requirement": min(min_requirement, max_limit),
                "Max Requirement": min(max_requirement, max_limit)
            }
        else:
            per_meal_requirements[meal] = {
                "Min Requirement": min_requirement,
                "Max Requirement": max_requirement
            }

    return per_meal_requirements

# Example for Calcium with daily requirement of 1000 mg and max limit of 2500 mg
daily_calcium_requirement = 50
calcium_max_limit = 100

calcium_per_meal = calculate_nutrient_requirements(daily_calcium_requirement, "Calcium", calcium_max_limit)

import json
print(json.dumps(calcium_per_meal, indent=4))
```

```python
# Calculate nutrient per meal


def calculate_nutrient_per_meal(cart_items):
    # Initialize a dictionary to store the total value for each nutrient
    total_nutrients = {}

    # Iterate over each dish in the cart
    for dish in cart_items:
        nutrients = dish["ingredient_distributed_nutrients"]

        # For each nutrient category (e.g., macro_nutrients, minerals, etc.)
        for category, nutrient_list in nutrients.items():
            for nutrient in nutrient_list:
                name = nutrient["name"]
                value = nutrient["value"]

                # Sum the nutrient values across all dishes
                if name in total_nutrients:
                    total_nutrients[name] += value
                else:
                    total_nutrients[name] = value

    return total_nutrients


# Sample Cart Data (Items in the order)
cart_items = [Masala_dosa,Biriyani,Paneer_Tikka,Mendu_Vada]

output = calculate_nutrient_per_meal(cart_items)

import json
print(json.dumps(output, indent=4))
```

```python
# Calculate nutrient fulfillment


def calculate_nutrient_fulfillment(order_nutrients, nutrients_data):

    fulfillment_results = []

    # Build a dictionary for the min_daily_intake values based on the nutrients_data
    min_intake_dict = {nutrient["name"]: nutrient["min_daily_intake"] for nutrient in nutrients_data}
    rating_dict = {nutrient["name"]: nutrient["rating"] for nutrient in nutrients_data}

    # Iterate through each nutrient in the order
    for nutrient, order_amount in order_nutrients.items():

        # Check if the nutrient exists in the min_intake_dict
        if nutrient in min_intake_dict:

            # Get the required amount for the current nutrient per meal
            meal_amount = min_intake_dict[nutrient]
            rating = rating_dict[nutrient]
            # print(meal_amount + "     " +order_amount)
            # Calculate the percentage fulfillment
            if meal_amount > 0:
                fulfillment_percentage = (order_amount / meal_amount) * 100

            else:
                fulfillment_percentage = 0  # If no requirement, no fulfillment (can adjust logic as needed)


            # Store the result for this nutrient
            fulfillment_results.append({
                "name": nutrient,
                "rating": rating,
                "order_contains": order_amount,
                "meal_requirement": meal_amount,
                "fulfillment_percentage": round(fulfillment_percentage, 2)
            })
        else:

            # Handle case where the nutrient is not in the meal requirements
            fulfillment_results.append({
                "name": nutrient,
                "rating": rating,
                "order_contains": order_amount,
                "meal_requirement": None,
                "fulfillment_percentage": None
            })

    return fulfillment_results



# Call the function with the order and nutrients data
result = calculate_nutrient_fulfillment(output, nutrients_data)

# Sort the data by 'rating' in descending order
result = sorted(result, key=lambda x: x['rating'], reverse=True)

import json
print(json.dumps(result, indent=4))

```
